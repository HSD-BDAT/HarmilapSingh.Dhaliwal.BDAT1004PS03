```python
# Step 1. Import the necessary libraries  
import pandas as pd

# Step 2. Import the dataset from this address. https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user   
# Step 3. Assign it to a variable called users.  
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = "|")

# Step 4. Discover what is the mean age per occupation  
print("Mean ager per occupation:")
print(users.groupby('occupation')['age'].mean().round(2))
```

---
#### Step 5. Discover the Male ratio per occupation and sort it from the most to the least 

```python
print("Male ratio per occupation")
print(users.groupby(['occupation', 'gender'])['gender'].count().unstack().fillna(0).
      apply(lambda x:round((x*100)/sum(x), 2), 1).sort_values('M', ascending = False))
```

---
#### Step 6. For each occupation, calculate the minimum and maximum ages  

```python
print("Min. and Max. age each occupation")
print(users.groupby('occupation')['age'].agg(['max', 'min']))
```

---
#### Step 7. For each combination of occupation and sex, calculate the mean age  

```python
print("Mean age for occupation and sex")
print(users.groupby(['occupation', 'gender'])['age'].mean().round(2).unstack())
```

---
#### Step 8. For each occupation present the percentage of women and men

```python
print("\nGender ratio occupation wise")
print(users.groupby(['occupation', 'gender'])['gender'].count().unstack().fillna(0).
      apply(lambda x:round((x*100)/sum(x), 2), 1))
```

## Question 4 (Wind Statistics)

The data have been modified to contain some missing values, identified by NaN.
Using pandas should make this exercise easier, in particular for the bonus question.
You should be able to perform all of these operations without using a for loop or other looping construct.

The data in 'wind.data' has the following format:

The first three columns are year, month, and day. The remaining 12 columns are average windspeeds in knots at 12 locations in Ireland on that day.

 




Step 11. Find the average windspeed in January for each location.
Treat January 1961 and January 1962 both as January.  
Step 12. Downsample the record to a yearly frequency for each location.  
Step 13. Downsample the record to a monthly frequency for each location.  
Step 14. Downsample the record to a weekly frequency for each location.  
Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
# Step 1. Import the necessary libraries  
import pandas as pd
import datetime

# Step 2. Import the dataset from this address 
# (https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data) 
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', sep = ' ')

# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.  
df.Mo.fillna(1, inplace = True)
df.Dy.fillna(1, inplace = True)
df['date'] = df[['Yr','Mo','Dy']].apply(lambda x: '-'.join(pd.to_numeric(x, downcast = 'integer').values.astype(str)), axis="columns")
df.date = pd.to_datetime(df.date)

# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.  
df.date[np.where(df.date.dt.year == 2061)[0]] = df.date[np.where(df.date.dt.year == 2061)[0]] - pd.offsets.DateOffset(years=100)

# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].  
df.index = df.date
df.drop(['Yr', 'Mo', 'Dy', 'date'], axis = 1, inplace = True)

# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below. 
df.isna().sum()

# Step 7. Compute how many non-missing values there are in total.  
df.shape[0] - df.isna().sum()

# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
round((df.mean()).mean(), 2)

# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations 
# of the windspeeds at each location over all the days
df.agg(['min', 'max', 'mean', 'std']).round(2)

# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations 
# of the windspeeds across all the locations at each day.
a = df.groupby(df.index.day_name()).agg(['mean', 'min', 'max', 'std'])
index = pd.MultiIndex.from_tuples(a.columns, names=["first", "second"])
a.columns = index
round(a.groupby(level = 1, axis = 1).mean(), 2)
```

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.  
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,7)
plt.rcParams['figure.dpi'] = 300

marriages = pd.read_csv('input/us-marriages-divorces-1867-2014.csv')

fig, ax = plt.subplots()
ax.plot(marriages['Year'], marriages['Marriages_per_1000'], label = 'Marriages', linewidth = 2)
ax.plot(marriages['Year'], marriages['Divorces_per_1000'], label = 'Divorces', linewidth = 2)

plt.xticks(np.arange(min(marriages['Year']) + 3, max(marriages['Year']), 10))
plt.xlabel('Year', size = 16)
plt.ylabel('# Per 1000', size = 16)
ax.legend()
ax.set_title('US Marriages Divorces (1867-2014)', size = 20)
```
